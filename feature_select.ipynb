```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("cleaned_K8.csv", header = None, low_memory = False)  # process the data with pandas
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
pca = PCA(n_components=50)
df = pca.fit_transform(df) 
df.shape
```

```python
tsne_em = TSNE(n_components=2, perplexity=30.0, n_iter=10000, verbose=1).fit_transform(df)
```

```python
import seaborn as sns
```

```python
z = pd.DataFrame()
z["comp-1"] = tsne_em[:,0]
z["comp-2"] = tsne_em[:,1]

sns.scatterplot(x="comp-1", y="comp-2",
                data=z).set(title="T-SNE projection") 
```
